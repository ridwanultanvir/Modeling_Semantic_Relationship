```python
import matplotlib.pyplot as plt
import numpy as np

# Test accuracy values in percentage
models = ["Frozen Embeddings", "Fine-tuned Embeddings", "Scratch Embeddings"]
test_accuracies = [67.29, 68.00, 66.66]  # Convert decimal values to percentages

# Set figure size (slightly wider for clarity)
plt.figure(figsize=(7, 5))

# Define bar width for a professional look (slightly reduced)
bar_width = 0.35

# Use muted, research-friendly colors
bar_colors = ["#4c72b0", "#55a868", "#c44e52"]  # Blue, Green, Red (Subtle, professional shades)

# Create bar chart with slimmer bars
bars = plt.bar(models, test_accuracies, color=bar_colors, alpha=0.85, width=bar_width, edgecolor="black", linewidth=1)

# Annotate each bar with its exact percentage value
for bar, acc in zip(bars, test_accuracies):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.15, f"{acc:.2f}%", 
             ha='center', fontsize=11, fontweight="bold", color="black")

# Labels and title (professional font size)
plt.ylabel("Test Accuracy (%)", fontsize=12, fontweight="bold")
plt.xlabel("BiLSTM Model", fontsize=12, fontweight="bold")
plt.title("Test Accuracy Comparison Across Embedding Types", fontsize=14, fontweight="bold")

# Proper y-ticks (clear, readable, not too dense)
yticks = np.arange(50, 71, 2)  # Y-axis from 50% to 70% in steps of 2%
plt.yticks(yticks, fontsize=11)  # Slightly larger font for readability
plt.ylim(50, 70)  # Ensuring the y-axis stays focused

# Rotate x-axis labels slightly for better readability
plt.xticks(rotation=15, fontsize=11)

# Add refined, light gray grid lines like graph paper (subtle but visible)
plt.grid(axis="y", linestyle="--", linewidth=0.6, color="gray", alpha=0.6)

# Remove unnecessary chart borders (for a cleaner look)
plt.gca().spines["top"].set_visible(False)
plt.gca().spines["right"].set_visible(False)

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Test accuracy values in percentage (including Stacked BiLSTM)
models = ["Frozen Embeddings", "Fine-tuned Embeddings", "Scratch Embeddings", "Stacked BiLSTM\nFine-tuned Embeddings"]
test_accuracies = [67.29, 68.00, 66.66, 69.50]  # Updated with Stacked BiLSTM

# Set figure size (slightly wider for clarity)
plt.figure(figsize=(8, 5))

# Define bar width for a professional look
bar_width = 0.35

# Use muted, research-friendly colors
bar_colors = ["#4c72b0", "#55a868", "#c44e52", "#8172b3"]  # Blue, Green, Red, Purple

# Create bar chart with slimmer bars
bars = plt.bar(models, test_accuracies, color=bar_colors, alpha=0.85, width=bar_width, edgecolor="black", linewidth=1)

# Annotate each bar with its exact percentage value
for bar, acc in zip(bars, test_accuracies):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.15, f"{acc:.2f}%", 
             ha='center', fontsize=11, fontweight="bold", color="black")

# Labels and title
plt.ylabel("Test Accuracy (%)", fontsize=12, fontweight="bold")
plt.xlabel("BiLSTM Model", fontsize=12, fontweight="bold")
plt.title("Test Accuracy Comparison", fontsize=14, fontweight="bold")

# Proper y-ticks
yticks = np.arange(50, 72, 2)  # Adjusted to include 70%+
plt.yticks(yticks, fontsize=11)
plt.ylim(50, 72)  # Extended to fit Stacked BiLSTM result

# Rotate x-axis labels slightly for better readability
plt.xticks(rotation=15, fontsize=11)

# Add refined, light gray grid lines (subtle but visible)
plt.grid(axis="y", linestyle="--", linewidth=0.6, color="gray", alpha=0.6)

# Remove unnecessary chart borders
plt.gca().spines["top"].set_visible(False)
plt.gca().spines["right"].set_visible(False)

# Show the plot
plt.show()
```
